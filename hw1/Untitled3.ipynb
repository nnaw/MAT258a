```julia
using Convex
using ECOS
n=100
A=randn(n,n)
b=randn(n,1)
x=Variable(n)
problem = minimize(0.5*sum_squares(A * x - b)+norm(x,1))
solve!(problem)
#r=b-A*x.value
using PyPlot
plt[:hist](x.value,facecolor="white",bins=linspace(-2,2,100))
xlim(-2,2)
ylim(0,40)
z=linspace(-2,2,100)
plot(z,abs(z))
```
