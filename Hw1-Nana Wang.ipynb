```julia
using Convex
using ECOS
n=100
m=30
srand(10000)
A=randn(n,m)
b=randn(n,1)*0.5
srand(0)
#A=rand(n,m)
#b=rand(n,1)*2-1
x=Variable(m)
p= minimize(norm(A*x-b,1))
solve!(p, ECOSSolver())
```

```julia
r=b-A*x.value
using PyPlot
figure(figsize=(20,5))
z=-2:(0.1/50):2
plt[:hist](r,50,facecolor="w")
plot(z,10*abs(z),"k")
axis([-2,2,0,40])
```

```julia
x=Variable(m)
p= minimize(norm(A*x-b,2))
solve!(p, ECOSSolver())
r=b-A*x.value
using PyPlot
figure(figsize=(20,5))
z=-2:(0.1/50):2
plt[:hist](r,40,facecolor="w")
plot(z,3*z.^2,"k")
axis([-2,2,0,15])
```

```julia
x=Variable(m)
p= minimize(sum(max(0,abs(A*x-b)-0.5)))
solve!(p, ECOSSolver())
r=b-A*x.value
using PyPlot
figure(figsize=(20,5))
z=-2:(0.1/50):2
plt[:hist](r,40,facecolor="w")
plot(z,10*max(0,abs(z)-0.5),"k")
axis([-2,2,0,25])
```

```julia
x=Variable(m)
p= minimize(sum(-log(1-A*x+b)-log(1+A*x-b)),A*x-b>-1,A*x-b<1)
solve!(p, ECOSSolver())
```

```julia
r=b-A*x.value
using PyPlot
figure(figsize=(20,5))
z=-0.999999:(0.1/1000000):0.999999
z1=-2:(0.1/100):2
plt[:hist](r,40,facecolor="w")
plot(z,-2*log(1-z)-2*log(1+z),"k")
plot(z1,2*z1.^2,linestyle="--")
axis([-2,2,0,10])
```
